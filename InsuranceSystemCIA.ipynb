```python
class Customer:
    def __init__(self, customer_id, name, occupation, age):
        self.customer_id = customer_id
        self.name = name
        self.occupation = occupation
        self.age = age

class Policy(Customer):
    def __init__(self, customer_id, name, occupation, age, p_num, p_type, premium, duration):
        super().__init__(customer_id, name, occupation, age)
        if not isinstance(premium, (int, float)) or premium <= 0:
            raise ValueError("Invalid numeric input for premium amount.")
        self.policy_number = p_num
        self.policy_type = p_type
        self.premium_amount = premium
        self.duration = duration

class Agent:
    def __init__(self, agent_id, name, branch):
        self.agent_id = agent_id
        self.name = name
        self.branch = branch

class InsuranceSystem:
    def __init__(self):
        self.customers = {}
        self.policies = {}
        self.agents = {}

    def add_customer(self, cid=None, name=None, occ=None, age=None):
        if cid is None:
            cid = input("Enter Customer ID: ")
            name = input("Enter Name: ")
            occ = input("Enter Occupation: ")
            age = int(input("Enter Age: "))
        self.customers[cid] = Customer(cid, name, occ, age)
        print(f"Customer {name} registered.")

    def add_policy(self, cid=None, p_num=None, p_type=None, premium=None, duration=None):
        try:
            if cid is None:
                cid = input("Enter Customer ID: ")
                if cid not in self.customers:
                    raise KeyError("Customer does not exist.")
                p_num = input("Enter Policy Number: ")
                p_type = input("Enter Type: ")
                premium = float(input("Enter Premium: "))
                duration = input("Enter Duration: ")

            if cid not in self.customers:
                raise KeyError(f"Customer {cid} not found.")

            c = self.customers[cid]
            new_p = Policy(c.customer_id, c.name, c.occupation, c.age, p_num, p_type, premium, duration)
            self.policies[p_num] = new_p
            print(f"Policy {p_num} added successfully.")

        except ValueError as e: print(f"Error: {e}")
        except KeyError as e: print(f"Error: {e}")

    def update_policy(self):
        try:
            p_num = input("Enter Policy Number to update: ")
            if p_num not in self.policies:
                raise RuntimeError("Attempting to update policy record that does not exist.")
            new_val = float(input("Enter New Premium: "))
            self.policies[p_num].premium_amount = new_val
            print("Update complete.")
        except (RuntimeError, ValueError) as e: print(f"Error: {e}")

    def search_policy(self):
        p_num = input("Enter Policy Number: ")
        if p_num not in self.policies:
            raise KeyError("Policy not found.")
        p = self.policies[p_num]
        print(f"Details: {p.name} | {p.policy_type} | ${p.premium_amount}")

    def display_policies(self):
        for p in self.policies.values():
            print(f"Policy: {p.policy_number} | Holder: {p.name}")

    def add_agent(self):
        aid = input("Enter Agent ID: ")
        name = input("Enter Agent Name: ")
        loc = input("Enter Branch: ")
        self.agents[aid] = Agent(aid, name, loc)

    def display_agents(self):
        for a in self.agents.values():
            print(f"Agent: {a.name} | Branch: {a.branch}")

    def menu(self):
        while True:
            print("\n1.Add Customer 2.Add Policy 3.Update 4.Search 5.List Policies 6.Add Agent 7.List Agents 8.Exit")
            choice = input("Choice: ")
            if choice == '1': self.add_customer()
            elif choice == '2': self.add_policy()
            elif choice == '3': self.update_policy()
            elif choice == '4': self.search_policy()
            elif choice == '5': self.display_policies()
            elif choice == '6': self.add_agent()
            elif choice == '7': self.display_agents()
            elif choice == '8': break

sys = InsuranceSystem()
sys.add_customer("C1", "John", "IT", 30)
sys.add_policy("C1", "P99", "Life", 1200.0, "5 Years")

sys.menu()
```
